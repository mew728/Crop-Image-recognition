#MODEL


```python
inputs =keras.Input(shape=(224, 224, 3),name='Input')

con2D_h1=Conv2D(64,(3,3),activation='relu',name='con2D_h1')(inputs)
con2D_h2=Conv2D(64,(3,3),activation='relu',name='con2D_h2')(con2D_h1)
normzlization1 =BatchNormalization(name='normzlization1')(con2D_h2)
pooling1 =MaxPooling2D(pool_size=(2,2),name='pooling1')(normzlization1)

con2D_h3=Conv2D(128,(3,3),activation='relu',name='con2D_h3')(pooling1)
con2D_h4=Conv2D(128,(3,3),activation='relu',name='con2D_h4')(con2D_h3)
normzlization2 =BatchNormalization(name='normzlization2')(con2D_h4)
pooling2 =MaxPooling2D(pool_size=(2,2),name='pooling2')(normzlization2)

con2D_h5=Conv2D(256,(3,3),activation='relu',name='con2D_h5')(pooling2)
con2D_h6=Conv2D(256,(3,3),activation='relu',name='con2D_h6')(con2D_h5)
normzlization3 =BatchNormalization(name='normzlization3')(con2D_h6)
pooling3 =MaxPooling2D(pool_size=(2,2),name='pooling3')(normzlization3)

con2D_h7=Conv2D(512,(3,3),activation='relu',name='con2D_h7')(pooling3)
con2D_h8=Conv2D(512,(3,3),activation='relu',name='con2D_h8')(con2D_h7)
normzlization4 =BatchNormalization(name='normzlization4')(con2D_h8)
pooling4 =MaxPooling2D(pool_size=(2,2),name='pooling4')(normzlization4)

con2D_h9=Conv2D(512,(3,3),activation='relu',name='con2D_h9')(pooling4)
con2D_h10=Conv2D(512,(3,3),activation='relu',name='con2D_h10')(con2D_h9)
normzlization5 =BatchNormalization(name='normzlization5')(con2D_h10)
pooling5 =MaxPooling2D(pool_size=(2,2),name='pooling5')(normzlization5)

flatten =Flatten(name='flatten1')(pooling5)
dense1 =Dense(1024,activation='relu',name='dense1')(flatten)
normzlization7 =BatchNormalization(name='normzlization7')(dense1)
dense2 =Dense(1024,activation='relu',name='dense2')(normzlization7)
normzlization6 =BatchNormalization(name='normzlization6')(dense2)

output =Dense(14,activation='softmax',name='dense3')(normzlization6)

model=keras.Model(inputs=inputs,outputs=output)

opt = optimizers.Adam(learning_rate=0.0001)
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer=opt)
```

```python
# train_df, test_df= train_test_split(all, test_size=0.15 ,random_state=0)
all=shuffle(all)
batch_size = 32
image_size = (224,224)
# total_train=train_df.shape[0]
```

```python
# print(train_df.shape)
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.15 )

train_generator = train_datagen.flow_from_dataframe(
    all,
    '/content/gdrive/MyDrive/Colab Notebooks/Aidea/data/train/',
    x_col='path',
    y_col='label',
    target_size=image_size,
    class_mode='categorical',
    batch_size=batch_size,
    subset='training'
)
```

```python
validation_generator = train_datagen.flow_from_dataframe(
    all,
    '/content/gdrive/MyDrive/Colab Notebooks/Aidea/data/train/',
    x_col='path',
    y_col='label',
    target_size=image_size,
    class_mode='categorical',
    batch_size=batch_size,
    subset='validation'
)
```

```python
epochs=5
model.fit(
  train_generator,
  epochs=epochs,
  validation_data= validation_generator,
  # steps_per_epoch= total_train//batch_size,
  validation_steps = validation_generator.samples// batch_size,
  callbacks=callbacks
)
```
